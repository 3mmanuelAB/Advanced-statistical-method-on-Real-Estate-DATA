Multiple Linear Regression 

Given a real estate dataset. 

File: 'real_estate_price_size_year.csv'. 

To create a multiple linear regression. 

Using the dependent variable is 'price', while the independent variables are 'size' and 'year'.

# Importing the relevant libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import statsmodels.api as sm
sns.set()
```

# Loading the data

```python
edata = pd.read_csv("real_estate_price_size_year.csv")
edata
```

```python
edata.describe()
```

```python
edata.head()
```

# Creating the regression

# Declaring the dependent and the independent variables

```python
y = edata["price"]
x1 = edata[["size","year"]]
```

### Regression

```python
x = sm.add_constant(x1)
results = sm.OLS(y,x).fit()
results.summary()
```
