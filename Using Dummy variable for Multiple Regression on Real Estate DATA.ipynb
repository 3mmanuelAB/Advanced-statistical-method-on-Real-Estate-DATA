# Using Dummy variable or categorical predictors for Multiple Regression on Real Estate DATA


```python
# IMPORTING THE LIBRARIES
import numpy as np
import pandas as pd
import statsmodels.api as sm
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#LOADING THE REAL ESTATE DATA
data = pd.read_csv("real_estate_price_size_year_view.csv")
data
```

```python
data.head()
```

```python
data["view"] = data["view"].map({"Sea view":1,"No sea view":0})
data
```

```python
#DECLARING THE VARIABLE AS "PRICE" THE DEPENDENT AND "SIZE" AND "VIEW" AS INDEPENDENT VARIABLE
y = data["price"]
x1 = data[["size","view"]]
```

```python
# THE REGRESSION
x = sm.add_constant(x1)
results = sm.OLS(y,x).fit()
results.summary()
```

```python
plt.scatter(data["size"],y)
yhat_yes = 7.748e+04 + 218.7521*data["size"]
yhat_no = ((7.748e+04)+(5.756e+04)) + 218.7521*data["size"]
fig = plt.plot(data["size"],yhat_yes, c = "brown", lw =4, label ="regression line")
fig = plt.plot(data["size"], yhat_no, c ="yellow", lw =4, label ="regresion line")
plt.xlabel("size", fontsize = 30)
plt.ylabel("price", fontsize = 30)
plt.show()

```

```python
# TO MAKE PREDICTIONS WITH THIS REGRESSION, i will use a 
new_data = pd.DataFrame({'const': 1,'size': [1250, 970], 'view': [0, 1]})
new_data = new_data[['const','size','view']]
new_data
```

```python
#we use Paul and Janet as index who are prospective buyers
new_data.rename(index={0: 'Paul',1:'Janet'})
```

```python
predictions = results.predict(new_data)
# The result
predictions
```

```python
# Join the two data frames
# John and Janet who are prospective buyer will be able to get the properties at the predicted price respectively 
predictionsdf = pd.DataFrame({'Predictions':predictions})
joined = new_data.join(predictionsdf)
joined.rename(index={0: 'Paul',1:'Janet'})
```
